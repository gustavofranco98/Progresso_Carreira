```python
import pandas as pd
```

```python
dados = pd.read_csv('/content/bd_carreira.csv')
```

```python
qtd_cursos = len(dados)
qtd_cursos
```

```python
#Percentual dos temas base da Carreira em Análise de Dados#
dados['Tema'].value_counts(normalize = True) * 100
```

```python
#Percentual dos temas base da Carreira em Análise de Dados#
percentual_temas = dados['Tema'].value_counts(normalize = True) * 100
```

```python
#Tema com maior carga na Carreira#
dados['Tema'].mode()
```

```python
composicao_temas = dados['Tema'].value_counts()
```

```python
analise_temas = pd.DataFrame({'Composição_base': composicao_temas, 'Porcentagem (%)': percentual_temas})
```
